```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get('https://docs.python.org/3/library/dis.html')
#print(r)
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
#print(soup.prettify())
```

```python
instructions_section = soup.find('section', id='python-bytecode-instructions')
#print(instructions_section)

# printing line by line 
# lines = instructions_section.find_all('dl')
# for line in lines:
#     print(line.text)
```

```python
import pandas as pd
df = pd.DataFrame()
```

```python
 # for each dl take the class = sig-name descname as the title 
# and then <p> as the description
cols = ['title', 'description']
dat = pd.DataFrame(columns = cols)
lines = instructions_section.find_all('dl')

for line in lines:
    title = line.find(class_='sig-name descname')
    description = line.find('p')
    if description:
        d_clean = description.text
    else:
        d_clean = str(description)
    if title:
        t_clean = title.text
    else:
        t_clean = str(title)
    dat = dat.append({'title': (t_clean),'description':(d_clean)},ignore_index=True)
    
```
